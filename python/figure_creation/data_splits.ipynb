```python
import pandas as pd
import matplotlib.pyplot as plt

idx_to_label = {
    1: "Non-gesture",
    2: "Pointing with one finger",
    3: "Pointing with two fingers",
    4: "Click with one finger",
    5: "Click with two fingers",
    6: "Throw up",
    7: "Throw down",
    8: "Throw left",
    9: "Throw right",
    10: "Open twice",
    11: "Double click with one finger",
    12: "Double click with two fingers",
    13: "Zoom in",
    14: "Zoom out"
}
video_names = [idx_to_label[i] for i in range(2, 15)]

annotations_dir = "/mnt/RESOURCES/josemanuelbravo/GestureRecognitionSystem/data/IPN_Hand/annotations/"

test_annotations = pd.read_csv(annotations_dir + "Annot_TestList.txt")
train_annotations = pd.read_csv(annotations_dir + "Annot_TrainList.txt")
train_annotations = train_annotations[train_annotations["label"] != 1]
test_annotations = test_annotations[test_annotations["label"] != 1]
print("Train instances:", len(train_annotations))
print("Test instances:", len(test_annotations))
test_dist = test_annotations["label"].value_counts().sort_index()
train_dist = train_annotations["label"].value_counts().sort_index()
plt.figure(figsize=(11, 5))
plt.bar(range(0, 13), test_dist, zorder=3, label="Test", color='#f79645')
plt.bar(range(0, 13), train_dist, zorder=3, label="Train", bottom=test_dist, color='#45A5F7')
plt.title("Data distribution per class", weight='semibold')
plt.ylabel("Number of instaces")
plt.xticks(range(0, 13), video_names, rotation=70, size='small')
plt.legend()
plt.grid(zorder=0, alpha=0.3, linestyle='--')
```
